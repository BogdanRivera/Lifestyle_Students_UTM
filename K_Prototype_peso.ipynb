```python
import numpy as np
import pandas as pd
from kmodes.kprototypes import KPrototypes

Datos = pd.read_csv('Respuestas de formulario 1000 datos.csv')
```

```python
Datos
```

```python
Datos_df=Datos.drop(['Carrera', 'Obstaculos para realizar ejercicio', 'Lugares con comida balanceada'], axis=1)
```

```python
Datos_df
```

```python
SDatos_df = Datos_df.fillna(0)
array = Datos_df.values   #Se le asignan los valores de la variable a array
```

```python
array[:, 1] = array[:, 1].astype(float)  #Separa datos numericos
array[:, 2] = array[:, 1].astype(float)
array[:, 3] = array[:, 1].astype(float)
```

```python
## Remover outliers de edad

i=0
for x in array[:,1]:
    if x > 28:      ##No mayor a 28 años
        np.delete(array, i, axis=0)
        
    if x < 16:      ##No menor a 16 años
        np.delete(array, i, axis=0)
    i=i+1
```

```python
## Remover outliers de estatura

i=0
for x in array[:,2]:
    if x > 200:      ##No mayor a 2 metros
        np.delete(array, i, axis=0)
        
    if x < 140:      ##No menor a 1.4 metros
        np.delete(array, i, axis=0)
    i=i+1
```

```python
## Remover outliers de peso

i=0
for x in array[:,3]:
    if x > 110:      ##No mayor a 110 KG
        np.delete(array, i, axis=0)
        
    if x < 40:      ##No menor a 40 KG
        np.delete(array, i, axis=0)
    i=i+1
```

```python
array
```

```python
kproto = KPrototypes(n_clusters=3, verbose=2, max_iter=20)   #Se realiza el algoritmo K_Prototype
clusters = kproto.fit_predict(array, categorical = [0,4,5,6])
```

```python
print(kproto.cluster_centroids_)          #Se imprimen los centroides
```

```python
cluster_dict=[]                       #Se determina a que cluster pertenece cada dato
for c in clusters:
    cluster_dict.append(c)
```

```python
Datos['cluster']=cluster_dict           #Se agrea la columa Cluster a datos para identificar los datos de cada cluster
```
